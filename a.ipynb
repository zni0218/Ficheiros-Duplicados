```python
from pathlib import Path
import urllib.request
import os
```

```python
DATA_DIR = Path("./data/taxi_csv")

DATA_DIR.mkdir(parents=True, exist_ok=True)

print("Folder created:")
print(DATA_DIR.resolve())
```

```python
url_loc = {}

year = 2013

months = [1, 2, 3]

for m in months:

    month = f"{m:02d}"

    fname = f"yellow_tripdata_{year}-{month}.csv"

    url = f"https://s3.amazonaws.com/nyc-tlc/trip+data/{fname}"

    loc = DATA_DIR / fname

    url_loc[url] = loc
```

```python
opener = urllib.request.build_opener()

opener.addheaders = [
    (
        'User-Agent',
        'Mozilla/5.0'
    )
]

urllib.request.install_opener(opener)
```

```python
for url, loc in url_loc.items():

    try:

        print(f"\nDownloading:")
        print(url)

        urllib.request.urlretrieve(url, str(loc))

        print(f"Saved to:")
        print(loc)

    except Exception as e:

        print(f"ERROR downloading {url}")
        print(e)
```
